```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#assining target variable
```

```python
y=d1["Delivery Time"]
y
```

```python
x=d1[["Sorting Time"]]
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y,test_size=0.30,random_state=10)
```

```python
le=LinearRegression()
le.fit(x_train, y_train)
```

```python
y_pred=le.predict(x_test)
y_pred
```

```python
mse=mean_squared_error(y_test, y_pred)
```

```python
mse
```

```python
r2=r2_score(y_test, y_pred)
r2
```

```python
print("mean squared error:",mse)
print("R2 score:",r2)
```

```python
# Visualize the regression line
```

```python
plt.scatter(x_test, y_test, color='blue')
plt.plot(x_test, y_pred, color='red')
plt.xlabel("Sorting time")
plt.ylabel("Delivery time")
plt.title("Regression Line")
plt.show()
```

```python
# Predict delivery time for a sorting time of 5 hours
```

```python
sorting_time = np.array([[5]])
delivery_time_prediction = le.predict(sorting_time)
```

```python
print("Delivery time prediction:", delivery_time_prediction[0])
```

```python
# Apply Square Root Transformation
x_sqrt_train, x_sqrt_test = np.sqrt(x_train), np.sqrt(x_test)
le.fit(x_sqrt_train, y_train)
y_pred_sqrt = le.predict(x_sqrt_test)
mse_sqrt = mean_squared_error(y_test, y_pred_sqrt)
r2_sqrt = r2_score(y_test, y_pred_sqrt)
```

```python
# Print and visualize Square Root Transformation
print("\nModel with Square Root Transformation:")
print("Mean Squared Error:", mse_sqrt)
print("R2 Score:", r2_sqrt)
```

```python
plt.scatter(x_sqrt_test, y_test, color="Red")
plt.plot(x_sqrt_test, y_pred_sqrt, color="blue")
plt.xlabel("Square Root of Sorting Time")
plt.ylabel("Delivery Time")
plt.title("Linear Regression Line with Square Root Transformation")
plt.show()
```

```python
# Apply Logarithmic Transformation
x_log_train, x_log_test = np.log1p(x_train), np.log1p(x_test)
le.fit(x_log_train, y_train)
y_pred_log = le.predict(x_log_test)
mse_log = mean_squared_error(y_test, y_pred_log)
r2_log = r2_score(y_test, y_pred_log)
```

```python
# Print and visualize Logarithmic Transformation
print("\nModel with Logarithmic Transformation:")
print("Mean Squared Error:", mse_log)
print("R2 Score:", r2_log)
```

```python
plt.scatter(x_log_test, y_test, color="Red")
plt.plot(x_log_test, y_pred_log, color="blue")
plt.xlabel("Logarithm of Sorting Time")
plt.ylabel("Delivery Time")
plt.title("Linear Regression Line with Logarithmic Transformation")
plt.show()
```

```python
# Apply Exponential Transformation
x_exp_train, x_exp_test = np.exp(x_train), np.exp(x_test)
le.fit(x_exp_train, y_train)
y_pred_exp = le.predict(x_exp_test)
mse_exp = mean_squared_error(y_test, y_pred_exp)
r2_exp = r2_score(y_test, y_pred_exp)
```

```python
# Print and visualize Exponential Transformation
print("\nModel with Exponential Transformation:")
print("Mean Squared Error:", mse_exp)
print("R2 Score:", r2_exp)
```

```python
plt.scatter(x_exp_test, y_test, color="Red")
plt.plot(x_exp_test, y_pred_exp, color="blue")
plt.xlabel("Exponential of Sorting Time")
plt.ylabel("Delivery Time")
plt.title("Linear Regression Line with Exponential Transformation")
plt.show()
```

```python
# Apply Polynomial Transformation
x_poly_train, x_poly_test = x_train**2, x_test**2
le.fit(x_poly_train, y_train)
y_pred_poly = le.predict(x_poly_test)
mse_poly = mean_squared_error(y_test, y_pred_poly)
r2_poly = r2_score(y_test, y_pred_poly)
```

```python
# Print and visualize Polynomial Transformation
print("\nModel with Polynomial Transformation:")
print("Mean Squared Error:", mse_poly)
print("R2 Score:", r2_poly)
```

```python
plt.scatter(x_poly_test, y_test, color="Red")
plt.plot(x_poly_test, y_pred_poly, color="blue")
plt.xlabel("Square of Sorting Time")
plt.ylabel("Delivery Time")
plt.title("Linear Regression Line with Polynomial Transformation")
plt.show()
```

```python
'''Among the transformations applied, the logarithmic transformation yields the best results,
significantly improving the model's fit and capturing the underlying patterns in the data.
'''
```
